```python
import pandas as pd
df=pd.read_csv("finalclean.csv")
```

```python
import datetime
ts=df.groupby('Date')['indrtg'].count()
ts.index = pd.to_datetime(ts.index)
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
#plt.style.use('seaborn-darkgrid')
plt.style.use('fivethirtyeight')
matplotlib.rcParams.update({'font.size': 15})
fig,ax = plt.subplots(1,figsize=(18,9))
# fig=plt.figure(figsize=(18,8))
# curr_axes=plt.gca()
# cur_axes.axes.get_xaxis().set_visible(True)
ax.set_title('Review counts over time')
ts.plot(ax=ax)
#plt.savefig("time.png")
plt.show()
```

```python
dfselect=df[df['companyname']=='Blue Cross Blue Shield'].copy()
import datetime
#Positive Reviews
dfselectpos=dfselect[dfselect['indrtg']>=3].copy()
tspos=dfselectpos.groupby('Date')['indrtg'].count()
tspos.index = pd.to_datetime(tspos.index)

#Negative reviews
dfselectneg=dfselect[dfselect['indrtg']<3].copy()
tsneg=dfselectneg.groupby('Date')['indrtg'].count()
tsneg.index = pd.to_datetime(tsneg.index)

fig,ax = plt.subplots(1,figsize=(18,9))
# fig=plt.figure(figsize=(18,8))
# curr_axes=plt.gca()
# cur_axes.axes.get_xaxis().set_visible(True)

ax.set_title('Reviews for Blue Cross Blue Shield over time')
tspos.plot(ax=ax,label='Positive Reviews')
tsneg.plot(ax=ax,label='Negative Reviews')
legend = ax.legend(loc='upper center', shadow=True)
#plt.axis('off')
ax.grid(False)
ax.set_facecolor('xkcd:white')
plt.ylabel('Review Counts')
plt.xlabel('Time in Years')
plt.show()
```
